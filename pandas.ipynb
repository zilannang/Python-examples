### .apply()

Her tahılı "düşük", "orta" veya "yüksek" vitamin düzeyine sahip olarak sınıflandırmak istersek ne olur? Pekala, eğer elimizde sadece bir değer varsa, bunu yapacak bir fonksiyon yaratmak oldukça kolay olurdu.

```python
def get_vitamin_level(vitamins):
    if vitamins > 99:
        return 'high'
    elif vitamins > 24:
        return 'medium'
    else:
        return 'low'
```

```python
get_vitamin_level(100)
```

```python
get_vitamin_level(25)
```

Şimdi bu işlevi vitamin sütunundaki her değere uygulamak için `.apply()` kullanacağız.

```python
df.vitamins.apply(get_vitamin_level)    # bütün değerlere bu fonksiyonu uygula demek oluyor.
```

```python
df['vitamin_level'] = df.vitamins.apply(get_vitamin_level)
```

```python
df.head()
```

Ayrıca lambda fonksiyonlarını da kullanabilirsiniz.`.apply()`

```python
df['vitamin_perc'] = df.vitamins.apply(lambda x: str(x)+'%') #gelen değerin yanına '%' bunu ekle ve yaz
```

```python
df.head()
```

### .map()

Diyelim ki şu anda kodlanmış üretici sütununu üreticilerin gerçek adlarına dönüştürmek istiyoruz ve bize yardımcı olacak bu sözlüğümüz var.

```python
mfr_dict = {
    'A': 'Am Home Food Prods',
    'G': 'General Mills',
    'K': 'Kelloggs',
    'N': 'Nabisco',
    'P': 'Post',
    'Q': 'Quaker Oats',
    'R': 'Ralston Purina'
}
```

Tek yapmamız gereken bu sözlüğü mevcut üretici sütunumuzla eşleştirmek.

```python
df.mfr.map(mfr_dict)       # map şunu yapıyor; mfr deki her birini yap
```

```python
df['manufacturer'] = df.mfr.map(mfr_dict)
```

```python
df.head()
```

`.apply()`, `.map()`, ve `.applymap()` arasındaki farklar çok küçük olabilir. Daha fazlası için bu [referansa](https://www.geeksforgeeks.org/difference-between-map-applymap-and-apply-methods-in-pandas/) göz atın.

## Grouping and Aggregating Data - Verileri Gruplama ve Toplama

Son olarak, pandas içindeki verileri gruplandırma kavramına bakalım.

Her üretici için tahıl porsiyonu başına ortalama kalori nedir?

```python
df.groupby('manufacturer').calories.median()
```

```python
df.groupby('manufacturer').calories.median().sort_values(ascending=False)
```

`.groupby()` öğesine bir sütun listesi ileterek birden çok özniteliğe göre gruplandırabiliriz.

```python
df.groupby(['manufacturer', 'type'])[['calories']].median()
```

Birden çok sütuna göre gruplama, çok düzeyli bir indeksle sonuçlanır. Bunların sütun olmasını tercih edersek, indeksi sıfırlayabiliriz.

```python
df.groupby(['manufacturer', 'type'])[['calories']].median().reset_index()  # başına 0'dan başlayarak index koyar.
```

Birden çok sütunda toplama gerçekleştirebiliriz.

```python
df.groupby('manufacturer')[['calories', 'sugars']].median()
```

`.agg()` kullanarak ve bir toplama listesi ileterek aynı anda birden çok toplama yapabilirsiniz.

```python
df.groupby('manufacturer').calories.agg(['mean', 'std', 'median', 'count'])  # agg() ile birden fazla hesaplama yaptırır.
```

sütunlara farklı toplama işlevleri uygulamak için  `.agg()` öğesine bir sözlük iletin.

```python
(df.groupby('manufacturer')[['mfr','calories', 'sugars', 'rating']].agg({ 'mfr': 'count','calories': 'median', 'sugars': 'mean','rating': 'mean' }))
```
